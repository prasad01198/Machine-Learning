# Types of Graphs

## importing Library

```python
import pandas as pd
from matplotlib import pyplot as plt
```

## Creating a dummy dataframe

```python
data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Value1': [10, 15, 7, 10, 9],
    'Value2': [12, 8, 19, 14, 11]
}

df = pd.DataFrame(data)
df
```

## Line Plot

- <br> - A line plot shows data as a line. <br> - It's used to show how things change over time or categories

```python
plt.plot(df['Category'], df['Value1'])
plt.show()
```

## Scatter Plot

<br> - A scatter plot uses dots to represent values. <br> - It helps in showing the relationship between two variables.


```python
plt.scatter(df['Category'], df['Value2'])
plt.show()
```

## Bar Plot

<br> - A bar plot represents data with rectangular bars.<br> - It's good for comparing categories.

```python
plt.bar(df['Category'], df['Value1'])
plt.show()
```

## Histogram

<br> - A histogram shows how data is distributed by grouping it into "bins." <br> - It's useful for seeing the frequency of values.

```python
plt.hist(df['Value1'])
plt.show()
```

## Pie Chart

<br> - A pie chart shows proportions. <br> - Each slice represents a part of the whole.

```python
plt.pie(df['Value1'])
plt.show()
```

## Box Plot

<br> - A box plot helps visualize the spread and centers of data. <br> - It shows the minimum, first quartile, median, third quartile, and maximum.

```python
plt.boxplot(df['Value1'])
plt.show()
```

## Area Plot

<br> - An area plot is like a line plot but fills the area below the line.

```python
plt.stackplot(df['Category'], df['Value1'])
plt.show()
```

## Heatmap

<br> - A heatmap shows data using colors to represent values.

```python
plt.imshow(df[['Value1', 'Value2']].values)
plt.show()
```

## Errorbar Plot

<br> - An errorbar plot shows variability of data using bars.

```python
plt.errorbar(df['Category'], df['Value1'], yerr=1)
plt.show()
```

## Stem Plot 

<br> - A stem plot draws vertical lines from a baseline to points, with markers on top.

```python
plt.stem(df['Category'], df['Value2'])
plt.show()
```

## Violin Plot

<br> - A violin plot shows the distribution of data like a box plot but with density.

```python
plt.violinplot(df['Value1'])
plt.show()
```

# Modifying the graphs

<br> - we plots the graphs earlier are basic graphs we can modify them like adding titles n all things as followed

## plt.title()

<br> - This method adds a title to the graph. <br> - It’s used when we want to describe what the graph represents.

```python
plt.plot(df['Category'], df['Value1'])
plt.title('Line Graph')
plt.show()
```

## plt.xlabel()

<br> - This method sets a label for the X-axis <br> - explaining what the values on the X-axis represent.

```python
plt.plot(df['Category'], df['Value1'])
plt.title('Line Graph')
plt.xlabel('xlable-values')
plt.show()
```

## plt.ylabel()

<br> - This method sets a label for the y-axis <br> - explaining what the values on the y-axis represent.

```python
plt.plot(df['Category'], df['Value1'])
plt.title('Line Graph')
plt.xlabel('xlable-values')
plt.ylabel('ylable-values')
plt.show()
```

## plt.plot() with label

<br> - This method creates a line plot. <br> - The label parameter is used to describe each line in the graph when there are multiple lines <br> - so it can be added to the legend.

```python
plt.plot(df['Category'], df['Value1'], label = 'line Value 1')
plt.plot(df['Category'], df['Value2'], label = 'line Value 2')
plt.title('Line Graph')
plt.xlabel('xlable-values')
plt.ylabel('ylable-values')
plt.show()
```

## plt.legend()

<br> - This method adds a legend to the graph <br> - explaining what each line (or other plotted element) represents.

```python
plt.plot(df['Category'], df['Value1'], label = 'line Value 1')
plt.plot(df['Category'], df['Value2'], label = 'line Value 2')
plt.title('Line Graph')
plt.xlabel('xlable-values')
plt.ylabel('ylable-values')
plt.legend()
plt.show()
```

## plt.grid()

<br> - This method adds a grid to the graph <br> - making it easier to read the values.

```python
plt.plot(df['Category'], df['Value1'], label = 'line Value 1')
plt.plot(df['Category'], df['Value2'], label = 'line Value 2')
plt.title('Line Graph')
plt.xlabel('xlable-values')
plt.ylabel('ylable-values')
plt.legend()
plt.grid()
plt.show()
```

## plt.figure()

<br> - This method sets the size of the figure <br> - In the example figsize=(8, 6) means the figure is 8 inches wide and 6 inches tall. <br> - This must be define at the starting

```python
plt.figure(figsize=(8, 6))
plt.plot(df['Category'], df['Value1'], label = 'line Value 1')
plt.plot(df['Category'], df['Value2'], label = 'line Value 2')
plt.title('Line Graph')
plt.xlabel('xlable-values')
plt.ylabel('ylable-values')
plt.legend()
plt.grid()
plt.show()
```

## plt.savefig()

<br> - This method saves the graph as an image file <br> - like a PNG. The filename is passed as a parameter.

```python
plt.figure(figsize=(8, 6))
plt.plot(df['Category'], df['Value1'], label = 'line Value 1')
plt.plot(df['Category'], df['Value2'], label = 'line Value 2')
plt.title('Line Graph')
plt.xlabel('xlable-values')
plt.ylabel('ylable-values')
plt.legend()
plt.grid()
plt.savefig('line_plot.png')
plt.show()
```

## plt.subplot()

<br> - This method allows us to create multiple subplots in one figure. <br> - It divides the figure into a grid, <br> - and we specify which subplot we are working on.

```python
plt.subplot(1, 2, 1)
plt.plot(df['Category'], df['Value1'], label='line Value 1')
plt.title('Line Graph for Value 1')
plt.xlabel('xlable-values')
plt.ylabel('ylable-values')
plt.legend()
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(df['Category'], df['Value2'], label='line Value 2', color='orange')
plt.title('Line Graph for Value 2')
plt.xlabel('xlable-values')
plt.ylabel('ylable-values')
plt.legend()
plt.grid()

plt.show()
```

## plt.plot() with color and marker

<br> - The color parameter changes the color of the line <br> - and the marker adds symbols at the data points.

```python
plt.figure(figsize=(8, 6))
plt.plot(df['Category'], df['Value1'], label = 'line Value 1', color = 'red', marker = 'o')
plt.plot(df['Category'], df['Value2'], label = 'line Value 2', color = 'black', marker = 'o')
plt.title('Line Graph')
plt.xlabel('xlable-values')
plt.ylabel('ylable-values')
plt.legend()
plt.grid()
plt.savefig('line_plot.png')
plt.show()
```

## plt.plot() with linestyle

<br> - The linestyle parameter changes how the line is drawn <br> - such as making it dashed. <br> '-' (solid line) <br> '--' (dashed line) <br>  '-.' (dash-dot line) <br> ':' (dotted line)

```python
plt.figure(figsize=(8, 6))
plt.plot(df['Category'], df['Value1'], label = 'line Value 1', color = 'red', marker = 'o', linestyle = '--')
plt.plot(df['Category'], df['Value2'], label = 'line Value 2', color = 'black', marker = 'o', linestyle = ':')
plt.title('Line Graph')
plt.xlabel('xlable-values')
plt.ylabel('ylable-values')
plt.legend()
plt.grid()
plt.savefig('line_plot.png')
plt.show()
```

- while working with pie chart there are some additional methods or parameters as followed

### labels

<br>- Adds labels to each slice of the pie chart to describe what each slice represents <br>-  we have to use 'labels' attribute in pie method where we can pass list of labels we want to assign to the pie chart

```python
plt.pie(df['Value1'], labels=['A','B','C','D','E'])
plt.show()
```

### autopct

<br>- Automatically adds the percentage of each slice as text on the chart <br>- we have to use autopct attribute which accept decimal points to display the percentages

```python
plt.pie(df['Value1'], 
        labels=['A','B','C','D','E'], 
        autopct='%1.1f%%')
plt.show()
```

### startangle

<br>- Rotates the pie chart to start the first slice at a given angle (in degrees).

```python
plt.pie(df['Value1'], 
        labels=['A','B','C','D','E'], 
        autopct='%1.1f%%',
        startangle=90)
plt.show()
```

### colors

<br>- Specifies custom colors for the slices.

```python
plt.pie(df['Value1'], 
        labels=['A','B','C','D','E'], 
        autopct='%1.1f%%',
        startangle=90,
        colors=['yellow', 'red', 'blue'])
plt.show()
```

### shadow

<br>- Adds a shadow beneath the pie chart to give it a 3D effect.

```python
plt.pie(df['Value1'], 
        labels=['A','B','C','D','E'], 
        autopct='%1.1f%%',
        startangle=90,
        colors=['yellow', 'red', 'blue'],
        shadow=True)
plt.show()
```

### counterclock

<br>- By default, the pie chart is drawn counterclockwise. <br>- Setting counterclock=False makes the pie chart draw clockwise.

```python
plt.pie(df['Value1'], 
        labels=['A','B','C','D','E'], 
        autopct='%1.1f%%',
        startangle=90,
        colors=['yellow', 'red', 'blue'],
        shadow=True,
        counterclock=False)
plt.show()
```

### wedgeprops

<br>- Customizes the properties of the slices (wedges), <br>- such as width, edge color, etc.

```python
plt.pie(df['Value1'], 
        labels=['A','B','C','D','E'], 
        autopct='%1.1f%%',
        startangle=90,
        colors=['yellow', 'red', 'blue'],
        # shadow=True,
        counterclock=False,
        wedgeprops={'edgecolor': 'black', 'linewidth': 1})
plt.show()
```

### textprops

<br>- Customizes the properties of the text labels in the pie chart, <br>- like font size or color.

```python
plt.pie(df['Value1'], 
        labels=['A','B','C','D','E'], 
        autopct='%1.1f%%',
        startangle=90,
        colors=['yellow', 'red', 'blue'],
        # shadow=True,
        counterclock=False,
        wedgeprops={'edgecolor': 'black', 'linewidth': 1},
        textprops={'fontsize': 8, 'color': 'blue'})
plt.show()
```

### pctdistance

<br>- Controls the distance of the percentage labels from the center of the pie.

```python
plt.pie(df['Value1'], 
        labels=['A','B','C','D','E'], 
        autopct='%1.1f%%',
        startangle=90,
        colors=['yellow', 'red', 'blue'],
        # shadow=True,
        counterclock=False,
        wedgeprops={'edgecolor': 'black', 'linewidth': 1},
        textprops={'fontsize': 8, 'color': 'blue'},
        pctdistance=0.5)
plt.show()
```

### radius

<br> -Controls the radius (size) of the pie chart.

```python
plt.pie(df['Value1'], 
        labels=['A','B','C','D','E'], 
        autopct='%1.1f%%',
        startangle=90,
        colors=['yellow', 'red', 'blue'],
        # shadow=True,
        counterclock=False,
        wedgeprops={'edgecolor': 'black', 'linewidth': 1},
        textprops={'fontsize': 8, 'color': 'blue'},
        pctdistance=0.5,
        radius=1.5)
plt.show()
```

### frame

<br>- Adds a rectangular frame around the pie chart.

```python
plt.pie(df['Value1'], 
        labels=['A','B','C','D','E'], 
        autopct='%1.1f%%',
        startangle=90,
        colors=['yellow', 'red', 'blue'],
        # shadow=True,
        counterclock=False,
        wedgeprops={'edgecolor': 'black', 'linewidth': 1},
        textprops={'fontsize': 8, 'color': 'blue'},
        pctdistance=0.5,
        radius=1.5,
        frame=True)
plt.show()
```
